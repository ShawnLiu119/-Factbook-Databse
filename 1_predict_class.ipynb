```python
def predict(img_path, model):
    img = image.load_img(img_path, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_batch = np.expand_dims(img_array, axis=0)
    img_preprocessed = preprocess_input(img_batch)
    prediction = model.predict(img_preprocessed)
    print(decode_predictions(prediction, top=3)[0])
```

```python
import numpy as np
import torch
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
import copy
import sys
import cv2
import os
```

```python
def create_data(img_size):
    Kane = '/content/drive/MyDrive/transfer_learning/sports-celebrity-images/KaneWilliamson/'
    Kobe = '/content/drive/MyDrive/transfer_learning/sports-celebrity-images/KobeBryant'
    Maria = '/content/drive/MyDrive/transfer_learning/sports-celebrity-images/MariaSharapova'
    Ronaldo = '/content/drive/MyDrive/transfer_learning/sports-celebrity-images/Ronaldo'
  
    Labels = {Kane:0, Kobe:1, Maria:2, Ronaldo:3}

    # Initializing list for storing data
    data = []
    labels = np.array([])

    for label in Labels:
        for ls in os.listdir(label):
            path = os.path.join(label, ls)
      
            #read images from path
            img = cv2.imread(path, cv2.IMREAD_COLOR)
            img = cv2.resize(img, (img_size, img_size))

            # Convert color from BGR to RGB
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

            # Adding data into data and labels list
            data.append(np.array(img))
            labels = np.append(labels, Labels[label])

    return np.array(data), labels
```

```python
data, labels = create_data(299)

print(f"Processed data with shape {data.shape}")
print(f"Label with shape {labels.shape}")
```

```python
#data preparation
# (1)normalization
# (2)augmentation (flip & adding noise)

data = data / np.max(data)
print(f"Min of data is {np.min(data)}")
print(f"Max of data is {np.max(data)}")

# Copy data variable and store in flip_data variable
flip_data = copy.deepcopy(data)

# random index to flip!
flip_idx = np.round(np.random.rand(flip_data.shape[0]))

# Flip it!
for idx in range(len(flip_idx)):
    
    # Check if flip_idx is 1 then flip it. On the other hands, i will add noise instead!
    if flip_idx[idx] == 1:
        flip_data[idx,:] = cv2.flip(flip_data[idx], -1)
    
    # Adding noise for flip_idx is 0
    else:
        flip_data[idx, :] = flip_data[idx,:] + np.random.rand(*flip_data[idx].shape)
        flip_data[idx, :] /= np.max(flip_data[idx,:])
```

```python
#function to show image
# Function for showing images
def show_img(dataset, label):
    # Randomly choose image index
    random_idx = np.random.randint(low=0, high=dataset.shape[0], size=10)

    # Initializing fig and ax variables
    fig, axs = plt.subplots(2, 5, figsize=(16,8))

    # Looping through each axs
    for i, ax in enumerate(axs.flatten()):

        # Showing image
        ax.imshow(dataset[random_idx[i],:], cmap='gray')
        ax.axis('off')
        
        # Set title for each image
        if label[random_idx[i]] ==0:
            ax.set_title("Kane Williamson")
        elif label[random_idx[i]] ==1:
            ax.set_title("Kobe Bryant")
        elif label[random_idx[i]] ==2:
            ax.set_title("Maria Sharapova")
        elif label[random_idx[i]] ==3:
            ax.set_title("Ronaldo")
    fig.tight_layout()
    plt.show()
```

```python
show_img(flip_data, labels)
```

```python
# Concat data with flip_data variable
data = np.concatenate((data, flip_data))

# Concat labels twice because we use same data twice
labels = np.concatenate((labels, labels))

print(data.shape)
```

```python
index = np.random.permutation(len(data))

data_shuffle = data[index]
label_shuffle = labels[index]
```

```python
#split data into train and test dataset
train_data, test_data, train_label, test_label = train_test_split(data_shuffle, label_shuffle,
                                                                 test_size=.2, stratify=labels,
                                                                 random_state=555)
print(f"Train data shape {train_data.shape}")
print(f"Test data shape {test_data.shape}")
```

```python
#https://www.tensorflow.org/guide/keras/transfer_learning typical transfer-learning work flow
def create_model():
    base_model = tf.keras.applications.Xception(
        weights='imagenet',  # Load weights pre-trained on ImageNet.
        input_shape=(299, 299, 3),
        include_top=False)  # Do not include the ImageNet classifier at the top.
    
    #freeze the base model
    base_model.trainable = False

    inputs = tf.keras.Input(shape=(299, 299, 3))
    # We make sure that the base_model is running in inference mode here,
    # by passing `training=False`. This is important for fine-tuning, as you will
    # learn in a few paragraphs.
    x = base_model(inputs, training=False)
    # Convert features of shape `base_model.output_shape[1:]` to vectors
    x = tf.keras.layers.GlobalAveragePooling2D()(x)
    # A Dense classifier with number of unit (multi-class calssification)
    x = tf.keras.layers.Dense(256, activation = 'relu')(x)
    x = tf.keras.layers.Dropout(0.3)(x)
    outputs = tf.keras.layers.Dense(4, activation='softmax')(x)
    model = tf.keras.Model(inputs, outputs)
    
    return model
```

```python
model_ts = create_model()
model_ts.compile(optimizer='adam', loss='sparse_categorical_crossentropy',metrics=['accuracy'])

history = model_ts.fit(
    train_data, 
    train_label,
    epochs=10, 
    validation_data= (test_data, test_label),
    callbacks=tf.keras.callbacks.EarlyStopping(monitor='loss', patience=3)
)
```

```python
test = '/content/drive/MyDrive/transfer_learning/sports-celebrity-images/Kobe.jpg'

img = image.load_img(test, target_size=(299, 299))
img_array = image.img_to_array(img)
img_batch = np.expand_dims(img_array, axis=0)
img_preprocessed = preprocess_input(img_batch)


result = np.argmax(model_ts.predict(img_preprocessed))
```
